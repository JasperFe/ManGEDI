```python
S2_folder = r"F:\Users\jfeyen\PhD\DATA\Sentinel\2019\Zomer"
```

```python
#Extracting Raster values, when files are splitted in multiple parts
def ExtractGedi(GediDF,S2_folder):
    from tqdm import tqdm
    import os
    import rasterio
    import rasterstats
    # Nieuwe DF aanmaken, met lege kolommen die de pixelwaarden per band zullen bevatten.
    gedi_S2_tot = {}
    #Banden met een nan waarde geven
    num = 0
    #CRS van GediDF overnemen
    GediDF_CRS = GediDF.crs
    #In wat volgt: loopen over de beschikbare S2-files, en waar enkel waarden invullen waar er nog een NaN-waarde is.
    for file in os.listdir(S2_folder):
        num = num+1
        print('Started with file ' + str(num) + ' of ' + str(len(os.listdir(S2_folder))))
        name = 'im_' + str(num)
        S2_file = S2_folder+'\\'+file
        gedi_S2_sol = {}

        with rasterio.open(S2_file) as src:
            bandnames = src.descriptions
            raster_crs = src.crs
            transform = src.transform
            for b in tqdm(range(len(bandnames))):
                band = bandnames[b]
                array = src.read(b+1)
                result = pd.DataFrame(rasterstats.zonal_stats(GediDF.geometry.to_crs(raster_crs),array, affine=transform ,stats = ['mean']))
                del array
                gedi_S2_sol[band]= result
        gedi_S2_tot[name]= gedi_S2_sol

    
    # Change data to a Geopandas frame
    
    # NaN's opvullen + finale samenzetten
    S2_tot = pd.DataFrame()
    gedi_S2= pd.DataFrame()
    for b in bandnames:
            gedi_S2[b] =  gedi_S2_tot['im_1'][b]['mean']
    S2_tot = gedi_S2
    
    for i in range(len(gedi_S2_tot)):
        im = 'im_' + str(i+1)
        for b in bandnames:
            #VIS_TEXT[b] = VISTEXT_tot[im][b]['mean']
            #Nu opvullen waar NA's zitten 
            S2_tot[b][S2_tot[b].isnull()] = gedi_S2_tot[im][b]['mean'].astype('float')

    return S2_tot
```
